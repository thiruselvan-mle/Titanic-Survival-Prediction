### Step 5: Survival by Passenger Class
Visualize survival count based on ticket class.

```python
plot_survival_by_pclass(df)
```

**Insight:** 
- 1st class passengers had the highest survival rate.

### Step 6: Survival by FamilySize
Check if passengers with family survived more than alone passengers.

```python
plot_survival_by_famsize(df)
```

**Insight:** 
- Passengers with 2–4 family members had higher survival rates.
- Extremely large or single passengers (alone) survived less.

### Step 7: Survival by IsAlone
Check if traveling alone affects survival.

```python
plot_survival_by_isalone(df)
```

**Insight:** 
- Alone passengers had lower survival chances.
- Traveling with family increased survival probability.

### Step 8: Age distribution
Check how age is distributed and how it affects survival.

```python
plot_age_distribution(df)
```

**Insight:**

### Step 9: Age vs Survival

```python
plot_age_survival(df)
```

**Insight:**

- Elderly passengers had lower survival rates.
- Most survivors were between ~20–40 years old.

### Step 10: Fare distribution
Visualize fare distribution and relationship with survival.

```python
plot_fare_distribution(df)
```

**Insight:**

### Step 11: Fare vs Survival

```python
plot_fare_vs_survival(df)
```

**Insight:**

### Summary
